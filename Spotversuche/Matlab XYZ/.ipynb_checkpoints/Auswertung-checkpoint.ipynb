```python
import scipy, scipy.io
import numpy

import matplotlib.pyplot as plt
%pylab inline
# This is the 3D plotting toolkit
from mpl_toolkits.mplot3d import Axes3D
```

```python
mat = scipy.io.loadmat('S1.mat')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = mat['X']
y = mat['Y']
z = mat['Z']

ax.plot_wireframe(x,y,z)
```
