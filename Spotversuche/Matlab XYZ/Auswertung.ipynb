```python
import scipy, scipy.io
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
%matplotlib notebook
# This is the 3D plotting toolkit
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.mlab import griddata
```

```python
mat = scipy.io.loadmat('S1.mat')
```

Zu viele Daten. Müssen ausgedünnt werden.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.ravel(mat['X'])
y = np.ravel(mat['Y'])
z = np.ravel(mat['Z'])

Z = np.reshape(z, (97,97))

X, Y = np.mgrid[0:100:1, 0:100:1]

grid_x = griddata(x,y,z, np.linspace(x.min(),x.max(),100), np.linspace(y.min(),y.max(),100), interp='linear')

ax.plot_surface(X,Y,grid_x, cmap=matplotlib.cm.coolwarm, linewidth=5, antialiased=False)
```
